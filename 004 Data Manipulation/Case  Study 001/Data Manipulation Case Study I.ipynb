# 6.6:- Create a series and reverse all lower case words

```python
def reverse(indi): 
  str = "" 
  for i in indi: 
    str = i + str
  return str

indi = 'india 1998', 'big country', np.nan
```

```python
print ("Original string: ", end="") 
print (indi)
```

```python
print("Reversed string: ", end="") 
reversed(indi)
```
